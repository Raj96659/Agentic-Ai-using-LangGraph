```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from langchain_groq import ChatGroq
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
llm = ChatGroq(
    model = 'llama-3.3-70b-versatile'
)
```

```python
class LLM_STATE(TypedDict):
    question : str
    answer : str
```

```python
def llm_qa(state: LLM_STATE) -> LLM_STATE:

    question = state['question']

    prompt = f'Answer the following question {question}'

    answer = llm.invoke(prompt).content

    state['answer'] = answer
    
    return state
```

```python
# create our graph
graph = StateGraph(LLM_STATE)

# Add nodes to graph
graph.add_node('llm_qa',llm_qa)

# Add edges to graph
graph.add_edge(START,'llm_qa')
graph.add_edge('llm_qa',END)
 
workflow = graph.compile()
```

```python
intial_state = {'question':'What is full form of JEE exam?'}
workflow.invoke(intial_state)

# if you want only answer use this.
# workflow.invoke(intial_state['answer'])
```
